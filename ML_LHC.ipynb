#Q2 Determine the Total Number of Unique Tags

```python
u_tags = []

for tags in lst:
  l = tags.split('<')
  for ut in l[1:]:
    if ut.lower() not in u_tags:
      u_tags.append(ut.lower())

u_tags, len(u_tags)
```

```python
freq = {}

for tags in lst:
  l = tags.split('<')
  for tag in l[1:]:
    freq[tag.lower()] = freq.get(tag,0)+1

print(freq)
```

```python
top25 = sorted(freq, key=freq.get, reverse=True)[:25]
top25
```

```python
k = []
v = []

def keyfunction(k):
    return freq[k]

for key in sorted(freq, key=keyfunction, reverse=True)[:25]:
    k.append(key)
    v.append(freq[key])
```

```python
plt.rcParams["figure.figsize"] = (30,20)
plt.plot(k, v)
plt.xlabel('Tags')
plt.ylabel('Frequency')
plt.show()
```

```python
k1 = []
v1 = []

def keyfunction(k):
    return freq[k]

for key in sorted(freq, key=keyfunction, reverse=True)[:500]:
    k1.append(key)
    v1.append(freq[key])

plt.rcParams["figure.figsize"] = (100,20)
plt.plot(k1, v1)
plt.xlabel('Tags')
plt.ylabel('Frequency')
plt.show()
```
